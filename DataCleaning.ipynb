```python
import pandas as pd

# Loading the dataset
file_path = 'artDataset.csv'
art_data = pd.read_csv(file_path)

# Removing the 'Unnamed: 0' column
art_data_cleaned = art_data.drop(columns=['Unnamed: 0'])

# Checking for missing values in the dataset
missing_values = art_data_cleaned.isnull().sum()

# Removing the entry with the missing artist name
art_data_cleaned = art_data_cleaned.dropna(subset=['artist'])

# Further cleaning the price column to ensure all possible formatting issues are handled
art_data_cleaned['price_numeric'] = pd.to_numeric(art_data_cleaned['price'].str.replace('[^\d.]', '', regex=True), errors='coerce')

# Displaying the cleaned data and the new price column
display(art_data_cleaned.head(), art_data_cleaned.info())

# Saving the cleaned dataset to a new CSV file
art_data_cleaned.to_csv('cleanedArtDataset.csv', index=False)
```
